```python
from fastai.vision.all import *

from PIL import Image
def type_of_morocco_food(name):
    # Dictionary mapping keywords to categories
    categories = {
    'Bastila': 'Bastila',
    'Batbout': 'Batbout',
    'Briouat': 'Briouat',
    'Chebakia': 'Chebakia',
    'Couscous': 'Couscous',
    'Fekkas': 'Fekkas',
    'Harcha': 'Harcha',
    'Harira': 'Harira',
    'Kaab el ghazal': 'Kaab el ghazal',
    'Karan': 'Karan',
    'Khringo': 'Khringo',
    'Loubia': 'Loubia',
    'Maakouda': 'Maakouda',
    'Matbucha': 'Matbucha',
    'Mechoui': 'Mechoui',
    'Msemen': 'Msemen',
    'Rfissa': 'Rfissa',
    'Seffa': 'Seffa',
    'Sellou': 'Sellou',
    'Sfenj': 'Sfenj',
    'Shakchouka': 'Shakchouka',
    'Tagine': 'Tagine',
    'Tanjia': 'Tanjia',
    'Baghrir': 'Baghrir'
}


    # Check each category
    for keyword, category in categories.items():
        if keyword in name:
            return category
    
    return 'MA3ARFTOUCH'  # Default case if no keywords match

learn = load_learner('./models/morocco-food-ml-by-nouredinekn-v1.pkl')
```

```python
path='test-food/Untitled.jpg'
img = Image.open(path)
img
```

```python

name,_,probs=learn.predict(path)
if max(probs).item()>0.7:
    print(f'Food\'type : {name}')
    print(f'Probability it\'s : {max(probs).item():.6f}')
else:
    print('Not found')
```
